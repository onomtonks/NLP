```python
from spacy.symbols import ORTH
nlp.tokenizer.add_special_case('gimme',
    [
      {ORTH:'gim'},
      {ORTH:'me'}
    ]
)

doc = nlp('gimme some sunshine,gimme some space')

tokens =[tokens.text for tokens in doc]

tokens
```

```python

transactions = "Tony gave two $ to Peter, Bruce gave 500 € to Steve"
token=nlp(transactions)
l1=[]
l2=[]
for tok in token:
  if tok.like_num:
    l1.append(tok.text)
  if tok.is_currency:
    l2.append(tok.text)
d={l2[0]:l1[0],l2[1]:l1[1]}
for i,k in d.items():
    print(i,k)

```
